```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
Slices_heating = [2850.9,0,4388.8,2472,2816.4]
Slices_cooling = [1330,576,3709.7,432.9,4295.2]
Labels = ["opaque","IG","Fenestration","INF","Dist.Losses"]
col = ['c','m','r','b','g']
plt.title("cooling")
plt.pie(Slices_cooling,labels=Labels,colors=col,startangle=90,shadow=True,explode=(0,0,0,0,0.1),autopct="%1.1f%%")
```

```python
plt.title("heating")
plt.pie(Slices_heating,labels=Labels,colors=col,startangle=90,shadow=True,explode=(0,0,0.1,0,0),autopct="%1.1f%%")
```
