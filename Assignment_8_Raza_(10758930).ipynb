Use the same procedures that we employed in the lesson and write function that read the diffuse radiation, the direct radiation, the SLF, and F_shd and the PXI and write them on the corresponding window_DF columns

```python
import os
import pandas as pd
import numpy as np
```

```python
RLF_Tables = "/content/drive/MyDrive/EETBS 2021-2022/RLF Tables"
```

```python
path_windows_DF = os.path.join(RLF_Tables,"windows.csv")
```

```python
windows_DF = pd.read_csv(path_windows_DF,sep=";",index_col=0,header=0)
```

```python
windows_DF
```

```python
path_SLF = os.path.join(RLF_Tables,"SLF.csv")
```

```python
SLF = pd.read_csv(path_SLF,sep=";",index_col = 0, header = 0)
```

```python
SLF
```

I consider latitude to be 45

```python
windows_DF.loc[:,"Latitude"] = 45
```

```python
windows_DF
```

```python
def SLF_finder(row):
  Direction = row["Direction"]
  value_SLF = SLF.loc[Direction,"45"]
  return value_SLF
```

```python
windows_DF.loc[:,"SLF"]=windows_DF.apply(SLF_finder,axis=1)
```

```python
windows_DF
```

```python
FF_shd = (windows_DF.loc[:,"SLF"]*windows_DF.loc[:,"Doh"]-windows_DF.loc[:,"Xoh"])/windows_DF.loc[:,"Height"]
windows_DF.loc[:,"Fshd"] = np.minimum(1,np.maximum(0,FF_shd))  



  
```

```python
windows_DF
```

```python
path_BeamIrradiance = os.path.join(RLF_Tables,"BeamIrradiance.csv")
BeamIrradiance = pd.read_csv(path_BeamIrradiance,sep=";",index_col = 0, header = 0)
BeamIrradiance
```

```python
path_DiffuesIrradiance = os.path.join(RLF_Tables,"DiffuseIrradiance.csv")
DiffuseIrradiance = pd.read_csv(path_DiffuesIrradiance,sep=";",index_col = 0, header = 0)
DiffuseIrradiance
```

```python
def ED_finder(row):
  Direction = row["Direction"]
  value_ED = BeamIrradiance.loc[Direction,"45"]
  return value_ED
```

```python
def Ed_finder(row):
  Direction = row["Direction"]
  value_Ed = DiffuseIrradiance.loc[Direction,"45"]
  return value_Ed
```

```python
windows_DF.loc[:,"ED"]=windows_DF.apply(ED_finder,axis=1)
```

```python
windows_DF.loc[:,"Ed"]=windows_DF.apply(Ed_finder,axis=1)
```

```python
windows_DF
```

```python
PXI = windows_DF.loc[:,"Tx"]*(windows_DF.loc[:,"Ed"]+(1-windows_DF.loc[:,"Fshd"])*windows_DF.loc[:,"ED"])
windows_DF.loc[:,"PXI"] = PXI
```
